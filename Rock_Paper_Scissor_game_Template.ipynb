### Description 
> #### In this game, the user gets the first chance to pick the option among Rock, paper and scissors. After that computer select from the remaining two choices(randomly), the winner is decided as per the rules.
- Rock vs paper-> paper wins
- Rock vs scissor-> Rock wins
- paper vs scissor-> scissor wins.

```python
import random
from PIL import Image
from IPython.display import display
import pygame
# display image of game
img1 =Image.open('images (2).png')
display(img1)
# rules
winning_rules = 'winning rules of the Rock, paper and scissors as following : \nRock vs paper-> paper wins \nRock vs scissor-> Rock wins\npaper vs scissor-> scissor wins. '
print(winning_rules)
choices = '\nEnter choice \n1 for Rock ,\n2 for paper ,\nand 3 for scissor '
print(choices)

choice_dic ={ 1 : 'Rock' , 2 : 'paper' , 3 :'scissor'}


# function to display image of each choice
def display_image (choice) :
    if choice == 1:
        img = Image.open('rock.jpg')
    elif choice == 2:
        img = Image.open('paper.jpeg')
    elif choice == 3:
        img = Image.open('scissor.png')
        
    new_size = (150, 150)
    img_resized = img.resize(new_size)
    return display(img_resized)


while True :
    user_counter = 0
    computer_counter = 0 
    for  i in range(3) :
        try :
            user_choice = int(input('please enter your choice  digit from these numbers{ 1 ,2 ,3 } \nuser turn'))
            print('user choice is ',choice_dic[user_choice])
            display_image(user_choice)
            
            computer_choice = random.choice([1 , 2 , 3])
            print('now its computer turn ....\ncomputer choice is' ,choice_dic[computer_choice])
            display_image(computer_choice)
            
            print(f'\033[1;32m{choice_dic[user_choice]} vs {choice_dic[computer_choice]}\033[0m')
            
            # who win in the round
            if user_choice == 1 and computer_choice == 2 :
                computer_counter +=1
                print(f'\033[1m{choice_dic[computer_choice ]} wins ==>  <== computer wins in this round ==>\033[0m')
            elif user_choice == 2 and computer_choice == 3 :
                computer_counter +=1
                print(f'\033[1m{choice_dic[computer_choice ]} wins ==>  <== computer wins in this round ==>\033[0m')
            elif user_choice == 3  and computer_choice == 1 :
                computer_counter +=1
                print(f'\033[1m{choice_dic[computer_choice ]} wins ==>  <== computer wins in this round ==>\033[0m')
            elif user_choice == 2 and computer_choice == 1 :
                user_counter += 1 
                print( f'\033[1m{choice_dic[user_choice ]} wins ==>  <== user wins in this round ==>\033[0m')
            elif user_choice == 3 and computer_choice == 2 :
                user_counter += 1 
                print( f'\033[1m{choice_dic[user_choice ]} wins ==>  <== user wins in this round ==>\033[0m')
            elif user_choice == 1 and computer_choice == 3 :
                user_counter += 1 
                print( f'\033[1m{choice_dic[user_choice ]} wins ==>  <== user wins in this round ==>\033[0m')
            elif user_choice == computer_choice : 
                print('\033[1mtie in this round\033[0m ')
                
            # if (user_choice - computer_choice) % 3 == 1 :
            #     user_counter += 1 
            #     print( f'\033[1m{choice_dic[user_choice ]} wins ==>  <== user wins in this round ==>\033[0m')
            # elif (user_choice - computer_choice) % 3 == 2 :
            #     computer_counter +=1
            #     print(f'\033[1m{choice_dic[computer_choice ]} wins ==>  <== computer wins in this round ==>\033[0m')
            # elif (user_choice - computer_choice) % 3 == 0 :
            #     print('\033[1mtie in this round\033[0m ')
        
        except ValueError :
            print(' invalid input please only choice number from this three numbers [1 ,2 ,3]')
        except KeyError :
            print(' invalid choice please only choice number from this three numbers [1 ,2 ,3]')
        except Exception as e :
            print(type(e).__name__)
            print(e)
    # who win in the match
    if user_counter > computer_counter :
        print('\n\033[1;31mthe user is winner in this match\033[0m')
        pygame.mixer.init()
        pygame.mixer.music.load('winfantasia-6912.mp3')
        pygame.mixer.music.play()
    elif user_counter < computer_counter :
        print('\n\033[1;31mthe computer is winner in this match\033[0m')
        pygame.mixer.init()
        pygame.mixer.music.load('game-over-160612.mp3')
        pygame.mixer.music.play()
    elif user_counter == computer_counter :
        print('\n\033[1;31mThe match was tied\033[0m')
            
    
    answer = input('\ndo you want play again yes or no')
    while not(answer.lower() in ['yes' ,'no']):
        answer = input('do you want play again yes or no')
    if answer.lower() == 'no':
        print("thanks for playing\ngood bye!!")
        break
```
